# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
#
import pandas as pd

# URL for the CSV file
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"

df_insurance_company = pd.read_csv(url)
df_insurance_company
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
df_insurance_company.head()


```

```python
# Display column names to verify them
print("\nColumn names in the dataset:")
df_insurance_company.columns



```

```python
# Ensure all columns are in lowercase and spaces are replaced with underscores for consistency
df_insurance_company.columns = df_insurance_company.columns.str.strip().str.lower().str.replace(' ', '_')

# Display updated column names
print("\nUpdated column names:")
df_insurance_company.columns
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(df_insurance_company.isnull().sum())
```

```python
#(filling with mode for categorical and mean for numerical)
df_insurance_company['gender'].fillna(df_insurance_company['gender'].mode()[0], inplace=True)
df_insurance_company['vehicle_class'].fillna(df_insurance_company['vehicle_class'].mode()[0], inplace=True)
```

```python
# Verify if the columns 'gender' and 'vehicle_class' exist before converting
if 'gender' in df_insurance_company.columns:
    df_insurance_company['gender'] = df_insurance_company['gender'].astype('category')
else:
    print("\nColumn 'gender' does not exist in the dataset.")

if 'vehicle_class' in df_insurance_company.columns:
    df_insurance_company['vehicle_class'] = df_insurance_company['vehicle_class'].astype('category')
else:
    print("\nColumn 'vehicle_class' does not exist in the dataset.")
```

```python
# Standardize formats (example: date formats)
if 'effective_to_date' in df_insurance_company.columns:
    df_insurance_company['effective_to_date'] = pd.to_datetime(df_insurance_company['effective_to_date'])
else:
    print("\nColumn 'effective_to_date' does not exist in the dataset.")

# Print updated info to verify changes
print("\nDataFrame info after cleaning:")
df_insurance_company.info()
```

```python
if 'effective_to_date' in df_insurance_company.columns:
    df_insurance_company['policy_age'] = (pd.to_datetime('today') - df_insurance_company['effective_to_date']).dt.days

# Display the first few rows to verify the new feature
print("\nFirst few rows after adding 'policy_age':")
print(df_insurance_company.head())

# Save cleaned data to a new CSV file
df_insurance_company.to_csv('cleaned_marketing_customer_analysis.csv', index=False)

print("\nCleaned data saved to 'cleaned_marketing_customer_analysis.csv'.")
```

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
import pandas as pd

# URL for the CSV file
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"

df_insurance_company = pd.read_csv(url)

# Ensure all columns are in lowercase and spaces are replaced with underscores for consistency
df_insurance_company.columns = df_insurance_company.columns.str.strip().str.lower().str.replace(' ', '_')

# Convert data types if necessary
df_insurance_company['effective_to_date'] = pd.to_datetime(df_insurance_company['effective_to_date'])

# Display the first few rows of the dataset
df_insurance_company.head()
```

```python
# Create pivot table for total revenue by sales channel
total_revenue_by_channel = df_insurance_company.pivot_table(
    values='total_claim_amount',
    index='sales_channel',
    aggfunc='sum'
).round(2)

# Display the pivot table
print("\nTotal Revenue by Sales Channel:")
total_revenue_by_channel.reset_index()
```

```python
# Create pivot table for average customer lifetime value by gender and education level
avg_clv_by_gender_education = df_insurance_company.pivot_table(
    values='customer_lifetime_value',
    index='gender',
    columns='education',
    aggfunc='mean'
).round(2)

# Display the pivot table
print("\nAverage Customer Lifetime Value by Gender and Education Level:")
avg_clv_by_gender_education
```

```python
# Extract month from the effective to date
df_insurance_company['month'] = df_insurance_company['effective_to_date'].dt.month

# Create a summary table for the number of complaints by policy type and month
complaints_by_policy_month = df_insurance_company.pivot_table(
    values='number_of_open_complaints',
    index=['month', 'policy_type'],
    aggfunc='sum'
).reset_index()

# Display the summary table in long format
print("\nNumber of Complaints by Policy Type and Month:")
complaints_by_policy_month
```
